### 🔍 ¿Qué debes observar en tu pantalla?

Una vez que empiece a correr, vas a ver el verdadero **"Cómputo Off-chain"** en acción. Tu terminal imprimirá algo como esto:

1. **El escaneo:** Verás una lista de 5 números aleatorios (ej. `[410, 415, 405, 420, 412]`). Esto simula al sensor NDIR físico midiendo el aire cada segundo.
2. **El cálculo (Off-chain):** Python calculará el promedio (ej. `412 ppm`) y el Hash criptográfico en tu computadora. ¡Esto costó 0 dólares en Gas!
3. **La primera actualización:** Como la blockchain está vacía al principio, el sensor dirá *"Cambio significativo detectado"* y pagará Gas para subir el dato. Verás un mensaje de éxito con el Hash de la transacción.
4. **El Ahorro (La Magia):** Deja el bot corriendo unos segundos más. En el siguiente ciclo, si el nuevo promedio es `415 ppm`, el bot se dará cuenta de que la diferencia con el último dato en la blockchain (`412 ppm`) es menor a 50 (nuestro umbral).
* En ese momento, imprimirá: **"💤 Sin cambios significativos... Ahorrando Gas. No se envía transacción."**
